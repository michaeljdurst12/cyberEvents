```python
#import dependencies
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
 
```

```python
#load in data and rename titles
redteam = pd.read_csv("redteam.txt")
redteam.rename(columns={'150885': 'time', 'U620@DOM1': 'user@domain', 'C17693':'source_computer','C1003':'destination_computer'}, inplace=True)
redteam.head()
```

```python
#Build the graph
G=nx.from_pandas_dataframe(redteam, 'source_computer', 'destination_computer')
```

```python
# Plot it
nx.draw(G, node_size=20)
plt.show()
```
